```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-1, 1, 100)
y_1 = 2 * x + 1
y_2 = x ** 2
plt.plot(x, y_1)
plt.plot(x, y_2)
plt.show()
```

```python
plt.figure()
plt.plot(x, y_1)
plt.show()

plt.figure()
plt.plot(x, y_2)
plt.show()
```

```python
plt.figure(figsize=(2, 2))
plt.plot(x, y_1)
plt.show()

plt.figure(figsize=(2, 2))
plt.plot(x, y_2)
plt.show()
```

```python
plt.plot(x, y_1, color="red", linewidth=1, linestyle="--")
plt.plot(x, y_2, color="blue", linewidth=5, linestyle="-")
plt.show()
```
