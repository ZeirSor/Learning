```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-3, 3, 100)
y_1 = 2 * x + 1
y_2 = x ** 2

# xy范围
plt.xlim((-1, 2))
plt.ylim((-2, 3))

# xy描述
plt.xlabel('i am x')
plt.ylabel('i am y')

plt.plot(x, y_1, color="red", linewidth=1, linestyle="--")
plt.plot(x, y_2, color="blue", linewidth=5, linestyle="-")
plt.show()
```

```python
new_tick = np.linspace(-1, 1, 11)
print(new_tick)
```

```python
x = np.linspace(-1, 1, 100)
y_1 = 2 * x + 1
y_2 = x ** 2

# xy描述
plt.xlabel('i am x')
plt.ylabel('i am y')

plt.plot(x, y_1, color="red", linewidth=1, linestyle="--")
plt.plot(x, y_2, color="blue", linewidth=5, linestyle="-")
plt.xticks(new_tick)
plt.yticks([-1, 0, 1, 2, 3],
           ['level 1', 'level 2', 'level 3', 'level 4', 'level 5'])
plt.show()
```

```python
x = np.linspace(-1, 1, 100)
y_1 = 2 * x + 1
y_2 = x ** 2

# xy描述
plt.xlabel('i am x')
plt.ylabel('i am y')

plt.plot(x, y_1, color="red", linewidth=1, linestyle="--")
plt.plot(x, y_2, color="blue", linewidth=5, linestyle="-")

new_tick = np.linspace(-1, 1, 11)
print(new_tick)

plt.xticks(new_tick)
plt.yticks([-1, 0, 1, 2, 3],
           ['level 1', 'level 2', 'level 3', 'level 4', 'level 5'])

# get current axis
ax = plt.gca()
# 把右边和上边的边框去掉
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
# # 把x轴的刻度设置为‘bottom’
# ax.xaxis.set_ticks_position('bottom')
# ax.yaxis.set_ticks_position('left')

ax.spines['bottom'].set_position(('data', 0))
ax.spines['left'].set_position(('data', 0))
plt.show()
```
